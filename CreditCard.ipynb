```python
avg_income_by_name = (
    df.groupby("Name")["Annual_Income"]
    .mean()
    .reset_index()
    .sort_values(by="Annual_Income", ascending=False)
)

avg_income_by_name
```

```python
plt.figure(figsize=(10,5))

ax = sns.barplot(
    data=avg_income_by_name,
    x="Name",
    y="Annual_Income",
    color="steelblue"   
)

ax.set_title("Average Annual Income by Customer Name", fontsize=14)
ax.set_xlabel("Customer Name")
ax.set_ylabel("Average Annual Income")
plt.xticks(rotation=45)


plt.show()
```

```python
sns.set_theme(style="whitegrid")
avg_income = df.groupby('Employment_Type', as_index=False)['Annual_Income'].mean()
plt.figure(figsize=(6,4))

ax = sns.barplot(
    data=avg_income,
    x='Employment_Type',
    y='Annual_Income',
    hue='Employment_Type',   
    palette='Set2',
    legend=False             
)

# Add value labels
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom'
    )

plt.title('Average Annual Income by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Average Annual Income')
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

```python
Further Analysis On Age Group
```

```python
age_group_metrics = df.groupby('Age_group', observed=False)[['Annual_Income', 'Estimated_Credit_Limit', 'Credit_Score']].mean().reset_index()
age_group_metrics
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

#Average Annual Income
sns.barplot(
    data=age_group_metrics,
    x='Age_group',
    y='Annual_Income',
    hue='Age_group', 
    palette='viridis',
    legend=False,
    ax=axes[0]
)
axes[0].set_title('Avg Annual Income by Age Group')
axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Annual Income')

# Average Credit Score
sns.barplot(
    data=age_group_metrics,
    x='Age_group',
    y='Credit_Score',
    hue='Age_group', 
    palette='tab10',
    legend=False,
    ax=axes[1]
)
axes[1].set_title('Avg Credit Score by Age Group')
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Credit Score')

#Average Credit Limit
sns.barplot(
    data=age_group_metrics,
    x='Age_group',
    y='Estimated_Credit_Limit',
    hue='Age_group', 
    palette='hls',
    legend=False,
    ax=axes[2]
)
axes[2].set_title('Avg Credit Limit by Age Group')
axes[2].set_xlabel('Age Group')
axes[2].set_ylabel('Credit Limit')

plt.tight_layout()
plt.show()
```
